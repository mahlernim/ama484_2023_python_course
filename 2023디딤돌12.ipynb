```python
# interactive python widgets 라이브러리를 이용한 실습
# Slider widget 2가지 활용
# 불필요한 출력을 막기 위해 마지막에 pass 추가
import ipywidgets as widgets

def calc_bmi(height, weight):
  bmi = weight / (height/100)**2
  print(round(bmi, 2))

heightSlider = widgets.IntSlider(value=170, min=100, max=200, description="키:")
weightSlider = widgets.FloatSlider(value=60, min=30, max=100, description="몸무게:")
widgets.interact(calc_bmi, height=heightSlider, weight=weightSlider)
pass
```

```python
# slider 대신 text widget 활용
heightText = widgets.IntText(value=170, description="키:")
weightText = widgets.FloatText(value=60, description="몸무게:")
widgets.interact(calc_bmi, height=heightText, weight=weightText)
pass
```

```python
# DatePicker widget 활용
# datetime library가 필요함. delta는 timedelta object이고 그 중 days attribute를 사용함
import datetime
def days_gap(start_date, end_date):
  delta = end_date - start_date
  print(delta)
  print(type(delta.days))
  print(f"과제 제출 기한이 {delta.days} 남았습니다.")

date1 = widgets.DatePicker(value=datetime.date.today())
date2 = widgets.DatePicker(value=datetime.date(2023, 5, 29))
widgets.interact(days_gap, start_date=date1, end_date=date2)
pass
```

```python
# 한 위젯의 값이 다른 위젯의 값에 영향을 미치는 예시
# observe에서 어느 위젯의 무슨 값이 변할 때 어느 함수를 실행시킬지 지정함.
# interact()를 사용하지 않는 경우 display()로 출력을 해줘야 함.
def update_a(self):
  bSlider.value = 100 - aSlider.value
def update_b(self):
  aSlider.value = 100 - bSlider.value

aSlider = widgets.IntSlider(value=50, min=0, max=100, description="A:")
bSlider = widgets.IntSlider(value=50, min=0, max=100, description="B:")

aSlider.observe(update_a, "value")
bSlider.observe(update_b, "value")

display(aSlider, bSlider)
```

```python
# 다양한 선택형 위젯을 이용
# options에 list만 지정하면 label과 value가 같은 값이 됨. tuple 형태로 (label, value)를 제공할 수 있음.
selection1 = widgets.Dropdown(
    options=[("큰거", 0), ("중간", 1), ("작은거", 2)],
    value=1,
    description="사이즈:",
)
selection2 = widgets.RadioButtons(
    options=[("페퍼로니", 0), ("슈프림", 1), ("불고기", 2)],
    description="토핑",    
)
selection3 = widgets.Select(
    options=["배달", "포장"],
    description="주문종류:",
)

def calculate_price(size, topping, delivery):
  size_price = [2, 1.5, 1]
  topping_price = [8000, 10000, 11000]
  price = int(size_price[size] * topping_price[topping])
  if delivery == "배달":
    price += 3000
  print("{:,}원".format(price))

widgets.interact(calculate_price, size=selection1, topping=selection2, delivery=selection3)
pass
```

```python
# 주사위를 굴렸을 때 홀수, 짝수가 얼마나 나오는지 시각화할 수 있는 코드 예시
# Checkbox 위젯은 체크가 되면 True, 아니면 False 값을 가짐(boolean)
import random
def roll_dice(times):
  results = []
  for i in range(times):
    results.append(random.randint(1,6))
  return results
def print_results(odd, even):
  global results
  result_text = ""
  for i in results:
    if i % 2 == 1:
      if odd == True:
        result_text += str(i) + " "
      else:
        result_text += "_ "
    else:
      if even == True:
        result_text += str(i) + " "
      else:
        result_text += "_ "
  print(result_text)
results = roll_dice(30)

oddCheck = widgets.Checkbox(description="홀수", value=True)
evenCheck = widgets.Checkbox(description="짝수", value=True)
widgets.interact(print_results, odd=oddCheck, even=evenCheck)

pass
```

```python
# Output widget을 이용해 영역을 나누는 예제
# click_action2에서 output1 내부를 모두 지운 뒤 버튼 2개를 새로 그려줌.
def click_action1(self):
  global clicks
  clicks += 1
  with output1:
    print("{}회 누름".format(clicks))
def click_action2(self):
  global clicks
  clicks = 0
  output1.clear_output()
  with output1:
    display(button1, button2)
output1 = widgets.Output(layout={"border":"1px solid black", "width":"300px"})
button1 = widgets.Button(description="눌러봐")
button1.on_click(click_action1)
button2 = widgets.Button(description="지우기")
button2.on_click(click_action2)
clicks = 0
with output1:
  display(button1)
  display(button2)
display(output1)

```

```python
# 주사위 굴리는 시뮬레이션 오브젝트를 DiceRoll 클래스를 이용해 생성
# 버튼을 클릭하면 rocknroll() 함수가 실행되고, (1)주사위를 굴리고 (2)어느 숫자가 몇 번 나왔는지 세고 (3)결과를 출력하는 일련의 작업이 실행됨.
import random
class DiceRoll:
  def __init__(self):
    self.counts = [0, 0, 0, 0, 0, 0]
    self.rolls = []
  def roll(self, times):
    for i in range(times):
      self.rolls.append(random.randint(1,6))
  def count(self):
    for i in range(6):
      self.counts[i] += self.rolls.count(i + 1)
    self.rolls = []
  def display(self):
    string = ""
    for i in range(6):
      string += "{}={:.1f}% ".format(i+1, self.counts[i] / sum(self.counts) * 100)
    print(string)
def rocknroll(self):
  diceRoll.roll(timesIntText.value)
  diceRoll.count()
  diceRoll.display()

diceRoll = DiceRoll()
timesIntText = widgets.IntText(value=10, display="주사위 굴릴 횟수:")
rollButton = widgets.Button(description="굴리기!")
rollButton.on_click(rocknroll)
display(timesIntText, rollButton)


```
